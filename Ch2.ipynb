```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('dds_datasets/dds_ch2_nyt/nyt1.csv')
data.describe()
```

```python
def age_group(age):
    if(age < 18): return '<18'
    elif(age > 18 and age < 25): return '18-24'
    elif(age >= 25 and age < 34): return '25-34'
    elif(age >= 35 and age < 44): return '35-44'
    elif(age >= 45 and age < 54): return '45-54'
    elif(age >= 55 and age < 64): return '55-64'
    elif(age >= 65): return '65+'

data['age_group'] = data.apply(lambda row: age_group(row['Age']), axis=1)
data.describe()
```

```python
data.head(20)
```

```python
# plot by age group
%matplotlib inline
sns.stripplot(x = "age_group",y = "Impressions", data = data, jitter = True)
```

```python
sns.swarmplot(x = "age_group",y = "Impressions", data = data)
```
