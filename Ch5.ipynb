```python
import pandas as pd
```

```python
data = pd.read_csv("dds_datasets/dds_ch5_binary-class-dataset.txt", delimiter='\t')
data.describe()
```

```python
data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
indep_cols = data.columns.values.tolist()
indep_cols.remove('y_buy')
print(indep_cols)
sns.pairplot(data, y_vars=["y_buy"], x_vars=indep_cols)
```

```python
# test-train split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import SGDClassifier
```

```python
# Spot-Check Algorithms
models = []
models.append(( 'LR' , LogisticRegressionCV()))
models.append(( 'KNN' , KNeighborsClassifier()))
models.append(( 'SGD' , SGDClassifier()))
# Test options and evaluation metric
num_folds = 10
seed = 7
scoring = 'roc_auc'
```

```python
X = data[indep_cols]
Y = data['y_buy']

validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y,
test_size=validation_size, random_state=seed)
```

```python
# evaluate each model in turn
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed)
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
print(results[0])
```

```python
# draw ROC 
from sklearn.metrics import roc_curve, auc
Y_score = LogisticRegressionCV().fit(X_train, Y_train).decision_function(X_validation)
fpr, tpr, _ = roc_curve(Y_validation, Y_score)
roc_auc = auc(fpr, tpr)
```

```python
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
